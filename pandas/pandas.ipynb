# Series

```python
import pandas as pd

a = [1,2,3,4]
var1 = pd.Series(a)

print(var1)
```

```python
b = [1,2,3,4]
var = pd.Series(b,dtype=float,index=['a','b','c','d'])
print(var)
```

```python
dic = {'Name':['Anand','Sanjay','Vaishu'],'RollNo':[65,56,57]}

var = pd.Series(dic)
print(var)
```

```python
s = pd.Series(12,index=[1,2,3,4,5,6,7])
print(s)
```

```python
s1 = pd.Series(12,index=[1,2,3,4,5,6,7])
s2 = pd.Series(12,index=[1,2,3,4])
print(s1+s2)
```

# DataFrame

```python
l = [1,2,3,4,5,6]
var = pd.DataFrame(l)

print(var)
print(type(var))
```

```python
a = {'a':[1,2,3,4,5],'b':[1,2,3,4,5]}

var = pd.DataFrame(a)

print(var)
print(type(var))
```

```python
a = {'a':[1,2,3,4,5],'b':[1,2,3,4,5],65:[1,2,3,4,5]}

var = pd.DataFrame(a)

print(type(var))
print(var)

print(var['a'][2])
```

```python
a = {'a':[1,2,3,4,5],'b':[1,2,3,4,5],65:[1,2,3,4,5]}

var = pd.DataFrame(a,columns=['b',65],index=['a','b','c','d','e'])

print(type(var))
print(var)

print(var['b'][0])
```

# merge

```python
a = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
b = pd.DataFrame({"A":[1,2,3,10],"c":[5,6,7,10]})
pd.merge(a,b)
```

```python
pd.merge(a,b,how="left")
```

```python
pd.merge(a,b,how="right")
```

```python
pd.merge(a,b,how="outer")
```

```python
pd.merge(a,b,how="inner")
```

```python
c=pd.DataFrame({"A":[1,2,3,4],"B":[2,3,4,5]})
d=pd.DataFrame({"A":[3,4,5,6],"B":[4,5,6,7]})

pd.merge(c,d,right_index=True,left_index=True,suffixes=("anand","vaish"))
```
